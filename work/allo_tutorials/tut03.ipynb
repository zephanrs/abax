## Data Types and Type Casting

```python
# import packages
import allo
from allo.ir.types import int16, int32, float32, Int, UInt, Float, Fixed
```

#### Implicit Casting

```python
def add(a: int32, b: int32) -> int32:
  # implicit casts to int33 to avoid overflow
  return a + b

s = allo.customize(add)
# notice implicit case in IR
print(s.module)
```

#### Explicit Casting

```python
def cast(a: int32) -> int16:
  # explicit cast:
  b: float32 = a
  c: float32 = b * 2
  d: float32 = float(a) * 2
  e: int16 = c + d
  return e

s = allo.customize(cast)
# see IR
print(s.module)
```
